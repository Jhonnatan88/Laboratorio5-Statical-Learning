# K-Means: Utilice el enfoque con las variables que quedaron como X para encontrar el numero de clusters mÃ¡s adecuado para agrupar los datos, recuerde que puede utilizar el metodo del codo para determinar cual es la cantidad de clusters adecuada, sin embargo en esta ocaciÃ³n debemos considerar cual es la cantidad de etiquetas colocadas correctamente a cada fila X y determinar el accuracy de la asignaciÃ³n.

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
# Normalizando la data
std_scaler = MinMaxScaler(feature_range = (0,1))
```

```python
x = std_scaler.fit_transform(x)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y ,test_size =0.2 ,random_state = 2022
                                                    , shuffle = True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 25)
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn import metrics
print("Entrenar el set de Certeza: ", metrics.accuracy_score(y_train, classifier.predict(x_train)))
print("Probar el set de Certeza: ", metrics.accuracy_score(y_test, y_pred))
```

# Aplique el mÃ©todo de PCA para realizar la clusterisaciÃ³n con el mismo criterio que se menciona en el inciso anterior.

```python
(df['status_type']).unique()
```

```python
video = df[df['status_type'] == 'video']
photo = df[df['status_type'] == 'photo']
link = df[df['status_type'] == 'link']
status = df[df['status_type'] == 'status']
video.shape, photo.shape, link.shape, status.shape
```

```python
video = video.sample(2*len(status), random_state=2022)
photo = photo.sample(2*len(status), random_state=2022)
video.shape, photo.shape, link.shape, status.shape
```

```python
new_df = video.append([photo], ignore_index=True)
new_df = new_df.append([link], ignore_index=True)
new_df = new_df.append([status], ignore_index=True)
new_df.shape
```

```python
new_df['status_type'] = new_df['status_type'].replace(['video'],'1')
new_df['status_type'] = new_df['status_type'].replace(['photo'],'2')
new_df['status_type'] = new_df['status_type'].replace(['status'],'3')
new_df['status_type'] = new_df['status_type'].replace(['link'],'4')
```

```python
y = new_df.loc[:,'status_type']
x = new_df.iloc[:, new_df.columns != 'status_type']
x = new_df.iloc[:, new_df.columns != 'status_published']
y.shape, x.shape
```

```python
x.dtypes
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, 
                                                    random_state=2022, shuffle=True)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)

x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)
```

```python
pca.explained_variance_ratio_
```

```python
np.sum(pca.explained_variance_ratio_)
```

```python
np.round(pca.components_, 2)
```

# Regresion Logsitica - PCA

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score

rlog_clf_pca = LogisticRegression(random_state=0)
rlog_clf_pca.fit(x_train_pca, y_train)
preds_rlog_pca = rlog_clf_pca.predict(x_test_pca)
accuracy_score(y_test, preds_rlog_pca)
```

# SVM - PCA

```python
from sklearn.svm import SVC 

svm_clf_pca = SVC()
svm_clf_pca.fit(x_train_pca, y_train)
preds_svm_pca = svm_clf_pca.predict(x_test_pca)
accuracy_score(y_test, preds_svm_pca)
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.decomposition import PCA
pcas = np.arange(2, 11)
accs = []

for pca in pcas:
    #Calculo PCA
    pca = PCA(n_components=pca)
    x_train_pca = pca.fit_transform(x_train)
    x_test_pca = pca.transform(x_test)
    
    #entrenamiento del modelo
    lgr_clf_pca = SVC()
    lgr_clf_pca.fit(x_train_pca, y_train)
    preds_lgr_pca = lgr_clf_pca.predict(x_test_pca)
    acc = accuracy_score(y_test, preds_lgr_pca)
    accs.append(acc)

plt.plot(pcas, accs, "-o", color="red")
plt.title("PCA con SVM")
plt.xlabel("Componentes Principales")
plt.ylabel("ACC")
plt.show()
```
